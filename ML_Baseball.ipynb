```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

estimators = [
    ('scalor', StandardScaler()),
    ('clf', DecisionTreeClassifier())
]

pipe = Pipeline(estimators)
```

```python
pipe.set_params(clf__max_depth=2)
pipe.set_params(clf__random_state=13)
```

```python
X = dfpipe.drop(['연봉', '선수명', '팀명'], axis=1)
y = dfpipe['연봉']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
X_train
```

```python
pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = pipe.predict(X_train)
y_pred_test = pipe.predict(X_test)

print('Train acc :', accuracy_score(y_train, y_pred_tr))
print('Test acc :', accuracy_score(y_test, y_pred_test))
```

##  통산 기록 클릭.

```python
baseball_teams = ['SSG', '롯데', "키움", 'LG', '두산', 'KT', 'KIA', '삼성', '한화', 'NC']
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from tqdm import tqdm_notebook
```

```python
URL = "https://www.koreabaseball.com/Player/Search.aspx"
driver = webdriver.Chrome("./chromedriver.exe")
driver.get(URL)
```

```python
# 2015년 부터 2020년 투수 연봉 가져오기
base_2nd = []

for teams in tqdm_notebook(baseball_teams):
    driver.find_element_by_class_name("select02").send_keys(teams)
    time.sleep(2)
    driver.find_element_by_css_selector("#cphContents_cphContents_cphContents_ddlPosition").send_keys('투수')
    time.sleep(2)
    
    
    searchResult = driver.find_element_by_css_selector("#cphContents_cphContents_cphContents_udpRecord > div.inquiry > p > strong > span").text ## 검색기록 갯수
    time.sleep(2)
    cnt = 0
    j = 0
    while cnt <= int(searchResult):
        j += 1
        cnt += 1
        if j == 20:
            j = 1
        if cnt >= 40: # 총개수가 모두 60개 미만이므로 이렇게 제한
            driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ucPager_btnNo3').click()
            time.sleep(2)
        elif cnt >= 20:
            driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ucPager_btnNo2').click()
            time.sleep(2)
        driver.find_element_by_class_name("select02").send_keys(teams)
        time.sleep(2)
        driver.find_element_by_css_selector("#cphContents_cphContents_cphContents_ddlPosition").send_keys('투수')
        time.sleep(2)
        driver.find_element_by_css_selector(f"#cphContents_cphContents_cphContents_udpRecord > div.inquiry > table > tbody > tr:nth-child({j}) > td:nth-child(2) > a").click()
        source = driver.page_source
        soup = BeautifulSoup(source, "html.parser")
        entire = soup.find(class_="player_basic")
        base = entire.find_all("li")
        time.sleep(2)
        if base[7].text.split()[1][-2:] == '달러':
            yearMoney = base[7].text.split()[1][:-3]
        else:
            yearMoney = base[7].text.split()[1][:-2]  ## 연봉
        name = base[0].text.split()[1]
        driver.find_element_by_css_selector("#contents > div.sub-content > div.player_info > div.tab-depth2 > ul > li:nth-child(2) > a").click() # 통산 기록 클릭
        entire2 = soup.find("tbody") ## 통계창
        base2 = entire2.find("tr")
        if base2.text == '기록이 없습니다.':
            driver.back()
            time.sleep(2)
            driver.find_element_by_class_name("select02").send_keys(teams)
            time.sleep(2)
            driver.find_element_by_css_selector("#cphContents_cphContents_cphContents_ddlPosition").send_keys('투수')
            time.sleep(2)
            continue
        else:
            # driver.find_element_by_css_selector("#contents > div.sub-content > div.player_info > div.tab-depth2 > ul > li:nth-child(2) > a").click() # 통산 기록 클릭
            source = driver.page_source
            soup = BeautifulSoup(source, "html.parser")
            entire2 = soup.find("tbody") ## 통계창
            base2 = entire2.find_all("tr")
            tmp_pitcher = []
            x = -20
            for i in range(len(base2)):
                x += 20
                
                for k in base2:
                    for y in k:
                        for z in y:
                            if z != '\n':
                                tmp_pitcher.append(z)
                
                time.sleep(1)
                years = tmp_pitcher[0+x]
                team= tmp_pitcher[1+x]
                Avgfault= tmp_pitcher[2+x]
                match_num= tmp_pitcher[3+x]
                perfect_pitch= tmp_pitcher[4+x]
                perfect_win= tmp_pitcher[5+x]
                win = tmp_pitcher[6+x]
                lose= tmp_pitcher[7+x]
                save= tmp_pitcher[8+x]
                hold= tmp_pitcher[9+x]
                win_rate= tmp_pitcher[10+x]
                batPlayerNum= tmp_pitcher[11+x]
                inning= tmp_pitcher[12+x][:3]
                hitted= tmp_pitcher[13+x]
                hittedHomerun= tmp_pitcher[14+x]
                ball4= tmp_pitcher[15+x]
                hitbody= tmp_pitcher[16+x]
                strikeOut= tmp_pitcher[17+x]
                lossPoint= tmp_pitcher[18+x]
                LosePointSelf= tmp_pitcher[19+x]
                base_2nd.append({
                    '순위': cnt,
                    '선수명': name,
                    '팀명': team,
                    '평균자책점': Avgfault,
                    '경기': match_num,
                    # '완투':perfect_pitch ,
                    # '완봉': perfect_win,
                    '승리': win,
                    '패배': lose,
                    '세이브': save,
                    '홀드': hold,
                    '승률': win_rate,
                    # '타자수': batPlayerNum,
                    '이닝': inning,
                    '피안타': hitted,
                    '홈런': hittedHomerun,
                    '볼넷': ball4,
                    '사구': hitbody,
                    '삼진': strikeOut,
                    '실점': lossPoint,
                    '자책점': LosePointSelf,
                    '연도': years,
                    '연봉': yearMoney
                })

                driver.back()
                time.sleep(1)
            driver.back()
            # driver.find_element_by_class_name("select02").send_keys(teams)
            # time.sleep(1)
            # driver.find_element_by_css_selector("#cphContents_cphContents_cphContents_ddlPosition").send_keys('투수')
            # time.sleep(1)
                
        
    

df_base_2nd = pd.DataFrame(base_2nd)
df_base_2nd
```
