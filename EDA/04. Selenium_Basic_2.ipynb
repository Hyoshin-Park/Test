```python
import pandas as pd
import time
from selenium import webdriver
```

```python
url = "https://finance.naver.com/marketindex/"
driver = webdriver.Chrome("../driver/chromedriver.exe")
driver.get(url)
```

```python
# 환전 고시 환율

content = driver.find_elements_by_css_selector("#exchangeList > li > a")
content[0].find_element_by_css_selector(".value").text
```

```python
# 주요 뉴스 제목

news = driver.find_elements_by_css_selector("#content > div.section_news > div > ul > li:nth-child(3) > p > a")
news[0].text
```

```python
# 밑 페이지에 있는 매매 기준율

driver.find_elements_by_css_selector("body > div > table > tbody > tr")
```

## iframe

# selenium 을 이용한 방법

```python
# iframe 태그 지정

iframe = driver.find_element_by_css_selector("#frame_ex1")

```

```python
# iframe 태그 이동

driver.switch_to_frame(iframe)
```

```python
contents = driver.find_elements_by_css_selector("body > div > table > tbody > tr")
contents[0].find_element_by_css_selector(".sale").text
```

```python
from tqdm import tqdm_notebook
```

```python
datas = []

for content in tqdm_notebook(contents):
    title = content.find_element_by_css_selector(".tit > a").text
    sale = content.find_element_by_css_selector(".sale").text
    link = content.find_element_by_css_selector(".tit > a").get_attribute("href")
    datas.append({
        "title": title,
        "sale": sale,
        "link": link
    })
    time.sleep(2)

df = pd.DataFrame(datas)
df.to_excel("./sel_naver_finace1.xlsx", encoding="utf-8")
```

```python
driver.quit()
```

# requests 를 이용한 방법

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://finance.naver.com/marketindex/exchangeList.naver"
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, "html.parser")
print(soup.prettify())
```

```python
soup.select_one(".sale").text
```

```python
contents = soup.select("tbody > tr") # 부모 태그
contents[0].select_one(".sale").text
```
