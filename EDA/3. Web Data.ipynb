```python
from urllib.parse import urljoin

url_base = "https://www.chicagomag.com/"

# 필요한 내용을 담을 빈 리스트
# 리스트로 하나씩 컬럼을 만들고, DataFrame으로 합칠예정
rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all("div", "sammy") # soup.select(".sammy")

for items in list_soup:
    rank.append(items.find(class_="sammyRank").text)
    tmp_string = items.find(class_="sammyListing").text
    main_menu.append(re.split(("\n|\r\n"), tmp_string)[0])
    cafe_name.append(re.split(("\n|\r\n"), tmp_string)[1])
    url_add.append(urljoin(url_base, items.find("a")["href"]))
```

## 시카고 맛집 데이터 분석 - 하위페이지

```python
# requirements
import pandas as pd
from urllib.request import urlopen, Request
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("../data/03. best_sandwiches_list_in_chicago.csv", index_col=0)
df.tail()
```

```python
df["URL"][0]
```

```python
req = requests.get(df["URL"][0], headers={"user-agent":ua.ie})
html = req.content
soup_tmp = BeautifulSoup(html, "html.parser")
soup_tmp.find("p", "addy")
```

```python
# reqular expression
price_tmp = soup_tmp.find("p", "addy").text
price_tmp
```

```python
import re
re.split(".,", price_tmp)
```

```python
price_tmp = re.split(".,", price_tmp)[0]
price_tmp
```

```python
price_tmp= price_tmp[0]
```

```python
tmp = re.search("\$\d+\.(\d+)?", price_tmp).group()
price_tmp[len(tmp) + 2:]
```

```python
from tqdm import tqdm

price = []
address = []

for idx, row in tqdm(df.iterrows()):
    req = requests.get(row["URL"], headers={"user-agent":ua.ie})
    html = req.content
    soup_tmp = BeautifulSoup(html, "html.parser")
    
    gettings = soup_tmp.find("p", "addy").text
    price_tmp = re.split(".,", gettings)[0]
    tmp = re.search("\$\d+\.(\d+)?", price_tmp).group()
    price.append(tmp)
    address.append(price_tmp[len(tmp) + 2:])
    print(idx)
```

```python
price
```

```python
address
```

```python
df["Price"] = price
df["Address"] = address
df = df.loc[:,["Rank", "Cafe", "Menu", "Price", "Address"]]
df.set_index("Rank", inplace=True)
df.head()
```

```python
df.to_csv("../data/03. best_sandwiches_list_in_chicago2.csv", sep=",", encoding="utf-8")
```

## 시카고 맛집 데이터 지도 시각화

```python
# requirements

import folium
import pandas as pd
import numpy as np
import googlemaps
from tqdm import tqdm

```

```python
df = pd.read_csv("../data/03. best_sandwiches_list_in_chicago2.csv", index_col=0)
df.tail()
```

```python
gmaps_key = "AIzaSyClE2nbjG_7HnQxD37XNJhrnJlSafAKfqA"
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
lat = []
lng = []

for idx, row in tqdm(df.iterrows()):
    if not row["Address"] == "Multiple location":
        target_name = row["Address"] + ", " + "Chicago"
        
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get("geometry")
        
        lat.append(location_output["location"]["lat"])
        lng.append(location_output["location"]["lng"])

    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
len(lat), len(lng)
```

```python
df.tail()
```

```python
df["lat"] = lat
df["lng"] = lng
df
```

```python
mapping = folium.Map(location=[41.8781136, -87.6297982], zoom_start=11)

for idx, row in df.iterrows():
    if not row["Address"] == "Multiple location":
        folium.Marker(
            location=[row["lat"], row["lng"]],
            popup=row["Cafe"],
            tooltip=row["Menu"],
            icon=folium.Icon(
                icon="coffee",
                prefix="fa"

            )

        ).add_to(mapping)

mapping
```
